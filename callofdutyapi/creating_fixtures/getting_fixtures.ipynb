```python
import pandas as pd
```

```python
df=pd.read_csv('call_of_duty_guns.csv')
```

```python
column_names=df.columns.to_list()
column_names.remove(column_names[1])
```

```python

del df['type']
del df['name']
for x in range(len(column_names)):
    print("{")
    print('"model":"callofdutyapi.stat",')
    print(f'"pk": {x+1},')
    print('"fields":{')
    print(f'"name":"{column_names[x]}",')
    print('"unit":"NULL"')
    print("}")
    print("},")
```

```python
gun_type_df=df[['name','type']]
```

```python
for x in range(len(gun_type_df)):
    row=gun_type_df.iloc[x].to_list()
    print("{")
    print('"model":"callofdutyapi.gun",')
    print(f'"pk": {x+1},')
    print('"fields":{')
    print(f'"name":"{row[0]}",')
    print(f'"type":"{row[1]}"')
    print("}")
    print("},")
```

```python
df=pd.read_csv('call_of_duty_guns.csv')
```

```python
del df['type']
del df['name']

```

```python
column_names=df.columns.to_list()
total_count=1
for i in range(len(df)):
    for j in range(len(column_names)):
        print("{")
        print('"model":"callofdutyapi.gunstat",')
        print(f'"pk": {total_count},')
        print('"fields":{')
        print(f'"gun_id":{i+1},')
        print(f'"stat_id":{j+1},')
        print(f'"value":{df.iloc[i][j]}')
        print("}")
        print("},")
        total_count+=1
```
